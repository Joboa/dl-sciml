```python
from scipy.io import loadmat
import matplotlib.pyplot as plt
import numpy as np

data = loadmat("Input/burgers_shock.mat")
x = data['x']
t = data['t']
usol = data['usol']

# Create a meshgrid for x and y
T, X = np.meshgrid(t, x) # X, Y = (100, 256)
print(T.shape)

# Create the contour plot
fig, ax = plt.subplots(figsize=(15, 8))
contour = ax.contourf(T, X, usol, cmap='jet')

# Add a colorbar
cbar = plt.colorbar(contour)
cbar.set_label('usol values')

# Add labels and title
ax.set_xlabel('T-axis')
ax.set_ylabel('X-axis')
ax.set_title('Contour Plot of usol')

# Show the plot
plt.show()
```

```python
from scipy.io import loadmat
import matplotlib.pyplot as plt
import numpy as np
from scipy.interpolate import griddata

data = loadmat("Input/burgers_shock.mat")
x = data['x'].flatten()
t = data['t'].flatten()
usol = data['usol']

# Create a meshgrid for x and t
T, X = np.meshgrid(t, x)

# Interpolate to a finer grid
t_fine = np.linspace(t.min(), t.max(), 300)
x_fine = np.linspace(x.min(), x.max(), 300)
T_fine, X_fine = np.meshgrid(t_fine, x_fine)
usol_fine = griddata((T.flatten(), X.flatten()), usol.flatten(), (T_fine, X_fine), method='cubic')

# Create the contour plot
fig, ax = plt.subplots(figsize=(15, 8))
contour = ax.contourf(T_fine, X_fine, usol_fine, cmap='jet')

# Add a colorbar
cbar = plt.colorbar(contour)
cbar.set_label('usol values')

# Add labels and title
ax.set_xlabel('T-axis')
ax.set_ylabel('X-axis')
ax.set_title('Smooth Contour Plot of usol')

# Show the plot
plt.show()
```
