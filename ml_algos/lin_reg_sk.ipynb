```python
# General steps using sklearn regression models

import pandas as pd
import numpy as np
import sklearn
import matplotlib.pyplot as plt
from pandas.core.common import random_state
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

birds_data = pd.read_csv("data/birds.csv")
# birds_data.head()
# birds_data.describe()

# How salary varies with experience

plt.scatter(birds_data["nVisitsNestling"],
            birds_data["futureBehavior"],
            color="red")
plt.title("Future Behaviour vs Visits ")
plt.xlabel("Visits")
plt.ylabel("Future Behaviour")
plt.show()

# Test and Train dataset
x = birds_data.iloc[:, :1]
y = birds_data.iloc[:, 1:]

x_train, x_test, y_train, y_test = train_test_split(x, y,
                                                    test_size=0.2,
                                                    random_state=0)

# Using sklearn regression model
regressor = LinearRegression()
regressor.fit(x_train, y_train)

# Results prediction with the model
y_pred_test = regressor.predict(x_test)
y_pred_train = regressor.predict(x_train)

# Plot results based on the training set
plt.scatter(x_train, y_train, color="blue")
plt.plot(x_train, y_pred_train, color="red")
plt.title("Future Behaviour vs Visits ")
plt.xlabel("Visits")
plt.ylabel("Future Behaviour")
plt.legend(["Actual", "Predicted"])
plt.show()

# Plot results based on the testing set
plt.scatter(x_test, y_test, color="magenta")
plt.plot(x_train, y_pred_train, color="cyan")
plt.title("Future Behaviour vs Visits ")
plt.xlabel("Visits")
plt.ylabel("Future Behaviour")
plt.legend(["Actual", "Predicted"])
plt.show()

# Equation of the regression line
slope = regressor.coef_
intercept = regressor.intercept_

print(slope, intercept)
```
